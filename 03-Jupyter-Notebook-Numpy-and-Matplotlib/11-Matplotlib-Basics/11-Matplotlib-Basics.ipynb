### 11 ``matplotlib`` 基础

```python
import numpy as np 
x = np.linspace(0, 10, 100) #[0,10] 切100段
```

```python
x
```

```python
y = np.sin(x) #对于每一个x 求一个y值
```

```python
y
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.plot(x, y) #以x为横轴，y为纵轴 .plot只返回了一个对象 在.show()之前可以进行很多操作
```

```python
plt.show()
```

```python
siny = y.copy()
```

```python
cosy = np.cos(x)
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show() #自动选择了不同颜色
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
plt.show()
```

关于color参数：[https://matplotlib.org/2.0.2/api/colors_api.html](https://matplotlib.org/2.0.2/api/colors_api.html)

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="-.")
plt.show()
```

关于linestyle参数：[https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html#sphx-glr-gallery-lines-bars-and-markers-linestyles-py](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html#sphx-glr-gallery-lines-bars-and-markers-linestyles-py)

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlim(-5, 15) #坐标轴严格范围（两边没空白）
plt.ylim(0, 1)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.axis([-1, 11, -2, 2])#对两个轴同时进行调节 留1空间
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend() #加上图例
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("Welcome to matplotlib world!") #标题
plt.show()
```

### Scatter Plot 散点图

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color="red")
plt.show()
```

（使用场景区别）折线图：横轴是特征，纵轴是取值；散点图：横轴和纵轴都是特征。

```python
x = np.random.normal(0, 1, 10000) #生成均值为0 方差为1 的正态分布
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y, alpha=0.1) #不透明度
plt.show()
```
